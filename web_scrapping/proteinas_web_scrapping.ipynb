```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url_page = "https://listado.mercadolibre.com.do/proteina-whey"
```

```python
page = requests.get(url_page)
```

```python
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
tabla = soup.find_all('h2',{'class':"poly-box"})
print(tabla)
```

```python
titulos = soup.find_all('a',{"poly-component__title"})
titulos = [i.text for i in titulos]
print(titulos)
```

```python
urls = soup.find_all('a',{"poly-component__title"})
urls = [i['href'] for i in urls]
print(urls)
```

```python
urls[0]
```

```python
price = soup.find_all('span',{'class':"andes-money-amount__fraction"})
price = [i.text for i in price]
print(price)
```

```python
df = pd.DataFrame({'Nombre':titulos,'Precios':price,'URLS':urls})
```

```python
df.head()
```

valores nulos?

```python
df.isnull().sum()
```

No tenemos valores nulos.
Veamos a ver cuanto son los valores de nuestros datos.

```python
df.info()
```

```python
df['Nombre'] = df['Nombre'].astype(str)
```

```python
df['Precios'] = df['Precios'].str.replace(',', '').str.strip()
df['Precios'] = pd.to_numeric(df['Precios'], errors='coerce')
df['Precios'].head()
```

## Confirmamos si esta todo los valores estan bien

```python
def mejorar_titulo(Titulos):
    Titulos = Titulos.replace('- Kg - Kg', '')
    Titulos = Titulos.replace('- L A - L A', '')
    parts = Titulos.split()
    sabor = parts[3]
    cantidad = parts[4]
    precio = parts[-1]
    nuevo_titulo = f"Prote√≠na {parts[1]} {sabor} {cantidad} a ${precio}"
    return nuevo_titulo

df['titulo_mejorado'] = df['Titulos'].apply(mejorar_titulo)
df['titulo_mejorado'] = df['Titulos']
del df['titulo_mejorado']
df['Titulos']
```

```python
df.info()
```
